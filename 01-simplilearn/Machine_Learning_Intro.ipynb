## Types of Machine Learning

Depending on wether the algorithm can self-train and predict a condition or needs to indetify patterns to derive outcomes, it can be divided into:

1. **supervised learning** - algoritm uses the data developed in supervised environdments to deliver an output. Inputs and outputs are known:
  - linear regression
  - logistic regression
  - support vector machines
  - decision trees
  
  Examples of Supervised Learning:
  - prediction of temperature increase when year-wise temperature increase is known
  - prediction of cropy yield when season-wise crop quality deterioration is known
  - classification of waste when easte items(input) and the type of waste are known

  Supervised learning helps two actions:
  1. classification
  2. regression

2. **Unsupervised Learning** - no superviser is required to prepare data as outputs amay not be known for such algorithms. The program finds hidden patterns and recognizes their relations.
   
   Examples of Unsupervised Learning:
   - identification of anomalies over geografical landscapes
   - identification and grouping of crop rot
   - identification of user groups, based on commonalities

3. **Reinforced Learning** - the program learns from previous errors. Interpreter rewards the algorithm when it finds the correct solution. If the output is not proper, the algorithm reiterates until it finds a better result.
  Examples:
  - YouTube recommendations:
    - Program shows the list of songs available when a user searches for a particular song.
    - When user selects a specific song, the system trains itself to remember and deliver a similar result for future searches.
  - Other examples:
    - games where players can play with bots
    - spell check
    - autocorrect tools
    - search recommendation engines

## Machine Learning Pipeline

A machine learning (ML) pipeline is a series of sequential steps used to codify and automate ML workflows to produce ML models. It is an end-to-end construct that includes nad orchestrates:
1. Data extractions
2. Raw data input
3. Preprocessing
4. Features
5. Outputs
6. Model parameters
7. Model training
8. Deployment
9. Predicting outputs

- The term `pipeline` implies a one-way flow of data. But, `ML pipelines` are cyclical and iterative.
- every step in the sequence is repeated until a succesful algorithm is achieved.
```
Data preparation => Feature extraction (Model Training, Test Data, Model) => Prediction
```
- specific outcome is validated by the supervisor

## Machine Learning and Operations Professionals (MLOps)
- MLOps is a set of practices that combines:
  1. ML
  2. DevOps
  3. Data engineering
- it ensures reliable and efficient deployment and maintenance of ML models in production systems

- MLOps aims to:
  - Improve communication and collaboration between MLOps
  - Shorten and manage complete development life cycle
  - Ensure continuous delivery of high-quality predictive services

- **Phases of MLOps**
  - Design
  - Model Development
  - Operations

1. Design Phase
  - understanding the business and data, and then designs the ML-powered software
2. Model development phase
  - applicability of ML for the problem is verified by implementing proof of concept (POC).
3. Operations Phase
  - delivers the developed ML model in production

- all three MLOps phases are interconnected and influence one another
